<font face="微软雅黑" size=2 color=black > 常用线条类型：
<font face="微软雅黑" size=2color=black > 
https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot

颜色： 蓝色 - 'b' 绿色 - 'g' 红色 - 'r' 青色 - 'c' 品红 - 'm' 黄色 - 'y' 黑色 - 'k' 白色 - 'w' 
线： 直线 - '-' 虚线 - '--' 点线 - ':' 点划线 - '-.'

```python
#从-π--π线性切分256个值，包括π
X = np.linspace(-np.pi,np.pi,256,endpoint=True)
Sin = np.sin(X)
Cos = np.cos(X)
##dpi:每英寸80个像素点(dots per inch),类似于分辨率,facecolor:背景颜色，edgecolor:边框颜色
fig = plt.figure(figsize=(10,6),dpi=80,facecolor='white',edgecolor='green')
ax = fig.add_subplot(1,1,1)
ax.plot(X,Sin,linewidth=5.5,color='red',linestyle='--',label=u'sin',zorder=1)
ax.plot(X,Cos,linewidth=1,color='blue',linestyle='-',label=u'余弦',zorder=1.1)
```

<font face="微软雅黑" size=3 color=black> 限定坐标轴范围xlim：

```python
ax.set_xlim(1.1*X.min(),1.1*X.max())#xlim:x方向数轴范围
ax.set_ylim(-1.1,1.1)
```

<font face="微软雅黑" size=3 color=black> 设置刻度xticks：

```python
ax.set_xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi])#设置x方向刻度标签
ax.set_xticklabels([r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])#设置x方向刻度标签
ax.set_yticks([-1,-0.5,0,0.5,1])
#ax.set_yticklabels([-1,-0.5,0,0.5,1])
```

<font face="微软雅黑" size=3 color=black> 调整坐标轴的位置spines

```python
#ax = plt.gca()#gca get current axesx 获取当前的轴域 gcf() to get the current figure
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')#设置x轴为bottom，可选为top，bottom，both，default，none
#设置bottom轴的位置为y=0 ，可选为'data','axes','outward'，分别代表不同的坐标系，比如'axes',0则是在轴开始的位置移动0
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')#设置y轴为bottom，可选为left，right，both，default，none
ax.spines['left'].set_position(('data',0))
```

<font face="微软雅黑" size=3 color=black> 设置图例legend

```python
ax.legend(loc='upper left', frameon=False)
```

<font face="微软雅黑" size=3 color=black> 特殊点的标注annotate：

```python
t = 2*np.pi/3
ax.plot([t,t],[0,np.cos(t)], color ='blue', linewidth=1.5, linestyle="--")
ax.scatter([t,],[np.cos(t),], 50, color ='blue')
ax.annotate(r'$cos(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
             xy=(t, np.cos(t)), xycoords='data',
             xytext=(+10, +30), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

ax.plot([t,t],[0,np.sin(t)], color ='red', linewidth=1.5, linestyle="--")
ax.scatter([t,],[np.sin(t),], 50, color ='red')
#annotate:注释
ax.annotate(r'$sin(\frac{2\pi}{3})=\frac{1}{2}$',
           xy=(t, np.sin(t)), xycoords='data',
             xytext=(+10, +30), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
```

<font face="微软雅黑" size=3 color=black> 图层设置：

```python
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)#调整刻度标签的大小
    #bbox:bounding box 边界框
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65,zorder=2 ))#facecolor:背景颜色，edgecolor:边框颜色
plt.show()
```

<font face="微软雅黑" size=4 color=#1E90F > 子图的使用

<font face="微软雅黑" size=2 color=black> subplot

```python
# 使用子图，一张大图分成几个小图
fig2 = plt.figure()
x = np.linspace(0, 2 * np.pi, 50)
ax2 = fig2.add_subplot(2, 1, 1) # （行，列，活跃区）
ax2.plot(x, np.sin(x), 'r')
ax3 = fig2.add_subplot(2, 1, 2)
ax3.plot(x, np.cos(x), 'g')
plt.show()
```

<font face="微软雅黑" size=2 color=black> subplot2grid用法：  
https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.subplot2grid.html   
shape(x,y)：网格的大小   
loc(x,y):子图开始的位置(初始位置是(0,0))   
rowspan：沿行展开的数，也就是子图的宽   
colspan : 沿列展开的数，也就是子图的长

```python
ax1 =plt.subplot2grid((3, 3), (0, 0), colspan=3)
ax1.plot([1, 2], [1, 2])    # 画小图
ax1.set_title(u'测试tick',fontproperties=font)  # 设置小图的标题
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)
ax4 = plt.subplot2grid((3, 3), (2, 0))
ax5 = plt.subplot2grid((3, 3), (2, 1))
plt.show()
```

<font face="微软雅黑" size=4 color=#1E90F> 散点图与直方图：  

```python
# 简单的散点图
x = np.linspace(0, 2 * np.pi, 50)
y = np.sin(x)
plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

```python
# 彩色映射散点图，size是点的大小，colour是颜色
x = np.random.rand(1000)
y = np.random.rand(1000)
size = np.random.rand(1000) * 50
colour = np.random.rand(1000)
plt.scatter(x, y, size, colour)
plt.colorbar()
plt.show()
```

```python
# 直方图，50代表数据容器的个数，也就是条的个数
x = np.random.randn(1000)
#density=True是频率图，False是频数图
plt.hist(x, bins=50,color='red',density=False)
plt.xlabel('Smarts')  
plt.ylabel('Probability') 
#r:raw string原始字符串，不要转义
plt.title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')  
plt.show()
```
